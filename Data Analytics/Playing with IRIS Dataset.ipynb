**1) Import the Toy IRIS Dataset**

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
cols = iris.feature_names
```

```python
df = pd.DataFrame(data=iris.data,columns = cols)
df['Species'] = iris['target']
df
```

Observation - Using Pandas library to import DataFrame along with adding column names to the dataset.

**2) Print the Head and Tail of the Dataset (first 5 and last 5 entries)**

```python
new_df = df.head(5)
new_df= new_df.append (df.tail(5))
new_df
```

Observation - Printing the first and last 5 rows using head and tail

**3) Create a New Column named "Average" and find the Average of existing four features and store the same**

```python
df['Average'] =df.iloc[:,0:4].mean(axis=1)
df
```

Observation - Finding the average of four features using mean(average) and iloc(selecting partocular rows and columns from the dataset)

**4) Create a New Column named "Sum" and find the Sum of existing four features and store the same**

```python
df['Sum'] = df['sepal length (cm)'] + df['sepal width (cm)'] + df['petal length (cm)'] + df ['petal width (cm)']
df
```

Observation - Finding the sum of four features using simple operators 

**5) Check how many entires of "IRIS Setosa", "IRIS Virgnica" and "IRIS Versicolor" are present in the Dataset**

```python
df["Species"] = df["Species"].replace(0,"IRIS - SETOSA")
df["Species"] = df["Species"].replace(1,"IRIS - VERSICOLOR")
df["Species"] = df["Species"].replace(2,"IRIS - VERGINICA")
df
```

```python
print("-"*20)
df["Species"].value_counts()
```

 Observation - Mapping values to their names using Replace

**6) Find the Descriptive Statistics about all the Six Columns (existing four + two made-up features)**

```python
df.describe()
```

Observation - Finding the descriptive statistics of the dataframe like mean, standard deviation, maximum value, minimum value etc. using Describe function.

**7) Find the entries whose Sum Values are more than a User Input**

```python
num1 = float(input("Enter a number:"))
print()

df[df["Sum"]>num1]
```

Observation - Taking user's input and checking if the sum values are greater than user input using indexing

**8) Find the entries whose Sepal_Length is greater than Petal_Width**

```python
df[df["sepal length (cm)"]>df["petal width (cm)"]]
```

Observation - Using indexing to to print entries whose sepal_length is greater than petal_width
