## Seaborn

```python
import seaborn as sns
```

```python
sns.catplot(x ="Sex", hue ="Survived",
kind ="count", data = df)
```

Observation - It shows the number of male and females survied using catplot

```python
sns.stripplot(x='Sex', y='Age', data=df)
```

Observation - It shows the number of male and female along with their age using strip plot

```python
sns.boxplot(x='Sex', y='Age', data=df, hue="Survived")
```

Observation - It shows the number of males and females survived using boxplot

```python
sns.violinplot(x='Sex', y='Age', data=df)
```

Observation - It shows the age of the male and females present using violinplot

```python
sns.pairplot(df, hue='Sex')
```

Observation - It shows the pairplot of the full dataset
